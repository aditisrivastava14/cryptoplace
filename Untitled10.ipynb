```python
# Step 4: robust loading utility
import glob
import json

def robust_read(folder, pattern="*histor*"):
    # find likely file by pattern
    files = list(folder.glob(f"*{pattern}*"))
    if not files:
        files = list(folder.iterdir())
    for f in files:
        name = f.name.lower()
        try:
            if name.endswith('.csv'):
                df = pd.read_csv(f, low_memory=False)
                print(f"Loaded CSV: {f.name}")
                return df, f
            if name.endswith('.xlsx') or name.endswith('.xls'):
                df = pd.read_excel(f)
                print(f"Loaded Excel: {f.name}")
                return df, f
            if name.endswith('.json'):
                df = pd.read_json(f, lines=True)
                print(f"Loaded JSON: {f.name}")
                return df, f
        except Exception as e:
            print("Failed to load", f.name, ":", e)
    # fallback: try every file
    for f in folder.iterdir():
        try:
            if f.suffix=='.csv':
                return pd.read_csv(f, low_memory=False), f
        except:
            pass
    raise FileNotFoundError("Could not load dataset; please note the exact filename.")

# load trades (adjust pattern if necessary)
trades, trades_file = robust_read(CSV_DIR, pattern="histor")
sentiment, sent_file = robust_read(CSV_DIR, pattern="fear")
print("Trades file:", trades_file)
print("Sentiment file:", sent_file)
print("trades.shape:", trades.shape)
print("sentiment.shape:", sentiment.shape)
```

```python
# Load trades dataset
trades = pd.read_csv(CSV_DIR/"historical_trades_download")

# Load sentiment dataset
sentiment = pd.read_csv(CSV_DIR/"fear_greed_download", parse_dates=['date'])

print("Trades shape:", trades.shape)
print("Sentiment shape:", sentiment.shape)

# Quick peek
display(trades.head(3))
display(sentiment.head(3))
```

```python
def normalize_cols(df):
    df = df.copy()
    df.columns = [c.strip().lower().replace(" ", "_") for c in df.columns]
    return df

trades = normalize_cols(trades)
sentiment = normalize_cols(sentiment)

print("Trades columns:", trades.columns.tolist())
print("Sentiment columns:", sentiment.columns.tolist())
```

```python
trades['time_parsed'] = pd.to_datetime(trades['timestamp_ist'], format="%d-%m-%Y %H:%M", errors="coerce")
trades['trade_date'] = trades['time_parsed'].dt.date
```

```python
# Convert numeric columns
trades['execution_price'] = pd.to_numeric(trades['execution_price'], errors='coerce')
trades['size_tokens'] = pd.to_numeric(trades['size_tokens'], errors='coerce')
trades['size_usd'] = pd.to_numeric(trades['size_usd'], errors='coerce')
trades['closed_pnl'] = pd.to_numeric(trades['closed_pnl'], errors='coerce')

# Clean side column
trades['side'] = trades['side'].astype(str).str.lower().map({'buy':'Buy', 'sell':'Sell'}).fillna(trades['side'])

# Create profit/loss indicator
trades['is_profit'] = (trades['closed_pnl'] > 0).astype(int)

trades[['execution_price','size_usd','closed_pnl','side','is_profit']].head()
```

```python
import numpy as np

# per-trade features
trades['abs_pnl'] = trades['closed_pnl'].abs()
trades['pnl_per_usd'] = trades['closed_pnl'] / trades['size_usd']
trades['log_size_usd'] = np.log1p(trades['size_usd'])
trades['side_enc'] = (trades['side'] == 'Buy').astype(int)

trades[['closed_pnl','abs_pnl','pnl_per_usd','log_size_usd','side_enc']].head()
```

```python
# daily aggregates
daily = trades.groupby('trade_date').agg(
    total_pnl = ('closed_pnl','sum'),
    avg_pnl = ('closed_pnl','mean'),
    median_pnl = ('closed_pnl','median'),
    total_volume_usd = ('size_usd','sum'),
    n_trades = ('closed_pnl','count'),
    win_rate = ('is_profit','mean')
).reset_index()

print(daily.head())
```

```python
sentiment['date_only'] = pd.to_datetime(sentiment['date']).dt.date
daily = daily.merge(sentiment[['date_only','classification']], left_on='trade_date', right_on='date_only', how='left')
daily['classification'] = daily['classification'].fillna('Unknown')

daily.head()
```

```python
daily['classification_simple'] = daily['classification'].replace({'Extreme Fear':'Fear','Extreme Greed':'Greed'})
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")

# 1) PnL over time
plt.figure(figsize=(12,4))
plt.plot(pd.to_datetime(daily['trade_date']), daily['total_pnl'], label="Total PnL")
plt.title("Daily Total PnL")
plt.xlabel("Date"); plt.ylabel("PnL (USD)")
plt.legend(); plt.tight_layout(); plt.show()

# 2) Boxplot: total_pnl by sentiment
plt.figure(figsize=(6,4))
sns.boxplot(data=daily, x='classification_simple', y='total_pnl')
plt.title("Total PnL by Sentiment")
plt.show()

# 3) Win rate by sentiment
plt.figure(figsize=(6,4))
sns.barplot(data=daily, x='classification_simple', y='win_rate')
plt.title("Win Rate by Sentiment")
plt.show()
```

```python
from scipy.stats import mannwhitneyu

greed = daily[daily['classification_simple']=='Greed']['total_pnl'].dropna()
fear = daily[daily['classification_simple']=='Fear']['total_pnl'].dropna()

stat, p = mannwhitneyu(greed, fear, alternative='two-sided')
print("Mann-Whitney U test p-value:", p)
print("Median PnL Greed:", greed.median(), "| Fear:", fear.median())
```

```python
# save cleaned and aggregated data
daily.to_csv(CSV_DIR/"daily_aggregates.csv", index=False)
trades.to_csv(CSV_DIR/"trades_cleaned.csv", index=False)
print("Saved daily_aggregates.csv and trades_cleaned.csv")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, roc_auc_score

# Merge trades with sentiment
model_df = trades.merge(sentiment[['date_only','classification']],
                        left_on='trade_date', right_on='date_only', how='left')

# Simplify sentiment labels
model_df['classification_simple'] = model_df['classification'].replace({
    'Extreme Fear':'Fear','Extreme Greed':'Greed'
}).fillna('Unknown')

# Features & target
features = ['log_size_usd','execution_price','side_enc','classification_simple']
model_df = model_df.dropna(subset=['log_size_usd','execution_price','is_profit'])
X = model_df[features]
y = model_df['is_profit']

# Preprocess: scale numeric, one-hot encode classification
preproc = ColumnTransformer([
    ('num', StandardScaler(), ['log_size_usd','execution_price']),
    ('cat', OneHotEncoder(handle_unknown='ignore'), ['classification_simple']),
    ('pass', 'passthrough', ['side_enc'])
])

# Logistic Regression
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)

pipe_lr = Pipeline([('pre', preproc), ('clf', LogisticRegression(max_iter=1000))])
pipe_lr.fit(X_train, y_train)
y_pred = pipe_lr.predict(X_test)
y_proba = pipe_lr.predict_proba(X_test)[:,1]

print("Logistic Regression Report:")
print(classification_report(y_test, y_pred))
print("ROC AUC:", roc_auc_score(y_test, y_proba))

# Random Forest baseline
pipe_rf = Pipeline([('pre', preproc), ('clf', RandomForestClassifier(n_estimators=200, random_state=42))])
pipe_rf.fit(X_train, y_train)
y_pred_rf = pipe_rf.predict(X_test)
y_proba_rf = pipe_rf.predict_proba(X_test)[:,1]

print("\nRandom Forest Report:")
print(classification_report(y_test, y_pred_rf))
print("ROC AUC:", roc_auc_score(y_test, y_proba_rf))
```
